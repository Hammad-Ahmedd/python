```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
x=[1,2,3,4,5]
y=[10,11,12,13,14]
y1=[15,16,17,18,20]
y2=[21,22,23,24,25]



# Graph 1
plt.subplot(2, 1, 1)  # (rows, columns, graph number)
plt.plot(x, y1, color="blue")
plt.title("Graph 1")

# Graph 2
plt.subplot(2, 1, 2)
plt.plot(x, y2, color="red")
plt.title("Graph 2")

plt.tight_layout()
plt.show()
```

## Pie Chart in Matplotlib

```python
size=[40,30,15,15]
labels=["Apple","Bananas","Cherries","Dates"]

plt.pie(size,labels=labels,autopct='%1.1f%%')
plt.title("Fruits Pie Chart")
plt.show()
```

## Save Graph

```python
plt.pie([50,30,20],labels=["C++","Python","Javascript"],autopct='%1.1f%%')
plt.title("Language pie chart")
plt.savefig("language.png",dpi=300,bbox_inches='tight')
plt.show()
```
