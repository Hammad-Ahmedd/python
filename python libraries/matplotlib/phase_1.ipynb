```python
import matplotlib.pylab as plt
```

```python
plt.plot([1, 2, 3, 4, 5], [10, 20, 15, 30, 25])
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Hammad")
plt.show()
```

## Change line style, color & markers

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]
plt.plot(x,y,
         color='red',
         linestyle="--",
         marker='o',
         markersize=10,
         linewidth=3
         )
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Styled Line Graph")
plt.show()
```

## Bar Chart (plt.bar)

```python
names = ["Ali", "Sara", "John", "Ayesha"]
scores = [85, 92, 78, 88]

plt.bar(names,scores)
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Student Scores Bar Chart")
plt.show()
```

## Histogram

```python
marks = [45, 55, 60, 62, 70, 72, 73, 75, 80, 82, 85, 88, 90, 92, 95]

plt.hist(marks,bins=10)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Distribution of Student Marks")
plt.show()
```

## Scatter Plots (plt.scatter)

```python
sx = [1, 2, 3, 4, 5]
sy = [3, 7, 2, 9, 5]

plt.scatter(sx,sy,
            color='red',
            )
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Basic Scatter Plot")
plt.show()
```
