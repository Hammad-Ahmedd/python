```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
penguins=sns.load_dataset("penguins")
penguins.head()
```

```python
penguins["species"].value_counts()
```

## Scatterplot

```python
sns.set_style("whitegrid")
# set context use to show the size of the graph
sns.set_context("paper")
sns.scatterplot(data=penguins,x="bill_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.scatterplot(data=penguins,x="species",y="body_mass_g",hue="sex")
```

## Stripplot

```python
sns.stripplot(data=penguins,x="species",y="body_mass_g",hue="sex",dodge=True,jitter=True)
```

## Swarmplot 

```python
swarm=sns.swarmplot(data=penguins,x="species",y="bill_depth_mm",hue="sex")
plt.savefig("swarm_graph.png",dpi=300,bbox_inches="tight")
plt.show()
```

## Histplot

```python
sns.histplot(data=penguins,x="body_mass_g",hue="island")
sns.set_style("darkgrid")
sns.set_context("paper")
```

## Kdeplot

```python
sns.kdeplot(data=penguins, x="body_mass_g", hue="sex", fill=True)
```

## boxenplot

```python
sns.boxenplot(data=penguins,x="species",y="bill_length_mm",hue="sex")
```

## violinplot

```python
sns.violinplot(data=penguins,x="species", y="bill_depth_mm", hue="sex", split=True)
```

## pairplot

```python
sns.pairplot(data=penguins,hue="species")
```
