```python
import sklearn
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

## Lesson 1 Part(a)

```python
data = {
    'YearsExperience': [1, 2, 3, 4, 5],   # The Input
    'Salary': [30000, 40000, 50000, 60000, 70000] # The Answer
}
df=pd.DataFrame(data)
df
```

```python
X=df[["YearsExperience"]]
X
```

```python
y=df['Salary']
y
```

# Part(b)
## Linear Regression

```python
modal=LinearRegression()
modal.fit(X,y)
print("Training complete! The model has learned the pattern.")
```

## Part(c)

```python
prediction=modal.predict([[6],[7],[8],[9]])
prediction
```
