```python
import numpy as np
```

## 1️⃣ Creating Arrays

```python
arr = np.array([1, 2, 3])
print(arr)
```

### Common Array Creation

```python
print(f"Zeros {np.zeros(3)}")
print(f"Ones {np.ones(5)}")
print(f"Arange {np.arange(0, 10, 2)}")
print(f"linspace {np.linspace(0, 1, 5)}")
```

## 2️⃣ Array Shape & Reshaping

```python
a = np.array([1,2,3,4])
b = a.reshape(2,2)
print(b)
```

### Important attributes:

```python
arr=np.array([[1,2,3],[4,5,6]])
print(f"Shape: {arr.shape}")
print(f"Number of dimensions: {arr.ndim}")
print(f"Total elements: {arr.size}")
print(f"Data type: {arr.dtype}")
```

## 3️⃣ Indexing & Slicing

```python
a = np.array([10,20,30,40])
print(a[1])
print(a[1:3])
```

## 4️⃣ Mathematical Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * 2)
print(a.mean())
print(a.sum())
```

## 5️⃣ Broadcasting (Very Important 🔥)

```python
a = np.array([1,2,3])
print(a + 5)
```

## 6️⃣ Random Numbers

```python
np.random.rand(3)
np.random.randint(1,10,5)
```

## 7️⃣ Boolean Masking (Important for ML)

```python
a = np.array([10,20,30,40])
print(a[a > 20])
```

## 8️⃣ Matrix Multiplication

```python
a = np.array([[1,2],
              [3,4]])

b = np.array([[5,6],
              [7,8]])

print(np.dot(a,b))
```

## 9️⃣ Linear Algebra (For ML)

```python
np.linalg.det(a)
np.linalg.inv(a)
np.linalg.eig(a)
```
