### Creating random tensors

Random tensors are tensors of some arbitrary size which contain random numbers.

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(7) # set seed for reproducibility
random_1 = random_1.normal(shape=(3, 2))
random_2 = tf.random.Generator.from_seed(7)
random_2 = random_2.normal(shape=(3, 2))

# Are they equal?
random_1, random_2, random_1 == random_2
```

```python
tf.random.set_seed(42)
suffle=tf.constant([
    [1,2,3],
    [4,5,6],
    [7,8,9]
])
print(tf.random.shuffle(suffle,seed=10))
print(tf.random.shuffle(suffle,seed=50))
```

### Getting information from tensors

When dealing with tensors you probably want to be aware of the following attributes:
* Shape
* Rank
* Axis or dimension
* Size

```python
rank_4_tensor = tf.zeros(shape=[2, 3, 4, 5])
rank_4_tensor
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor))
print("Total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
```

```python
rank_4_tensor[:2,:3,:2,:3]
```

```python
# Create Rank 2 tensor
rank_2_tensor=tf.constant([[1,2],[4,5]])
# get the last item of the each row
rank_2_tensor[:,-1]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor=rank_2_tensor[...,tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
once_tensor = tf.ones(shape=(2,3,2))
# You can add an outer batch axis by passing axis=0:
expend_tensor=tf.expand_dims(once_tensor,axis=0)
print(expend_tensor)
print("Tensor shape",expend_tensor.shape)
print("Tensor dimension",expend_tensor.ndim)
```

```python
# You can add an outer batch axis by passing axis=0:
print("When axis is 0\n",tf.expand_dims(once_tensor,axis=0))
# The new axis location matches Python list.insert(axis, 1):
print("When axis is 1\n",tf.expand_dims(once_tensor,axis=1))
# Following standard Python indexing rules, a negative axis counts from the end so axis=-1 adds an inner most dimension:
print("When axis is -1\n",tf.expand_dims(once_tensor,axis=-1))
```

**Matrix multiplication**

In machine learning, matrix multiplication is one of the most common tensor operations.

There are two rules our tensors (or matrices) need to fulfil if we're going to matrix multiply them:

1. The inner dimensions must match
2. The resulting matrix has the shape of the outer dimensions

```python
tf.matmul(tensor,tensor)
```

```python
# Matrix mutliplication with Python operator "@"
tensor @ tensor
```

```python
tensor.shape
```

```python
# Create a tensor (3, 2) tensor
X = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])
# Create another (3, 2) tensor
Y = tf.constant([[7, 8],
                 [9, 10],
                 [11, 12]])

X, Y
```

```python
# Try to matrix multiply tensors of same shape
tf.matmul(X, Y)
```

```python
# Let's change the shape of Y
tf.reshape(Y, shape=(2, 3))
```

```python
X.shape, tf.reshape(Y, shape=(2, 3)).shape
```

```python
# Try to matrix multiply X by reshaped Y
X @ tf.reshape(Y, shape=(2, 3))
```

```python
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
tf.reshape(X, shape=(2, 3)).shape, Y.shape
```

```python
# Try change the shape of X instead of Y
tf.matmul(tf.reshape(X, shape=(2, 3)), Y)
```

```python
# Can do the same with transpose
X, tf.transpose(X), tf.reshape(X, shape=(2, 3))
```

```python
# Try matrix multiplication with transpose rather than reshape
tf.matmul(tf.transpose(X), Y)
```

📖 **Resource:** Info and example of matrix multiplication: https://www.mathsisfun.com/algebra/matrix-multiplying.html

**The dot product**

Matrix multiplication is also referrred to as the dot product.

You can perform matrix multiplication using:
* `tf.matmul()`
* `tf.tensordot()`
* `@`

```python
# Perform the dot product on X and Y (requires X or Y to be transposed)
tf.tensordot(tf.transpose(X), Y, axes=1)
```

```python
# Perform matrix mutliplication between X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2, 3)))
```

```python
# Check the values of Y, reshape Y and transposed Y
print("Normal Y:")
print(Y, "\n") # "\n" is for newline

print("Y reshaped to (2, 3):")
print(tf.reshape(Y, (2, 3)), "\n")

print("Y transposed:")
print(tf.transpose(Y))
```

```python
tf.matmul(X, tf.transpose(Y))
```

Generally, when performing matrix multiplication on two tensors and one of the axes doesn't line up, you will tranpose (rather than reshape) one of the tensors to get satisify the matrix multiplication rules.

### Aggregating tensors

Aggregating tensors = condensing them from multiple values down to a smaller amount of values.

```python
tensor_range=tf.range(start=2,limit=10,delta=1)
tensor_aggregation=tf.constant(tensor_range,shape=(2,4))
tensor_aggregation
```

```python
tf.reduce_max(tensor_aggregation)
```

```python
tf.reduce_min(tensor_aggregation)
```

```python
tf.reduce_min(tensor_aggregation,axis=0)
```

```python
tf.reduce_min(tensor_aggregation,axis=1)
```

```python
tensor_mean=tf.reduce_mean(tensor_aggregation)
tensor_mean
```

```python
tf.reduce_mean(tensor_aggregation[:,:3],axis=0)
```

```python
tf.reduce_mean(tensor_aggregation,axis=0,keepdims=False)
```

```python
tf.reduce_mean(tensor_aggregation,axis=0,keepdims=True)
```

```python
tensor_aggregation_sum=tf.math.reduce_sum(tensor_aggregation,axis=0,keepdims=True)
tensor_aggregation_sum.numpy(),tensor_aggregation_sum.ndim
```

```python
small_tensor=tf.constant([[1,2,3,6]],dtype=tf.float16)
tf.math.reduce_variance(small_tensor,axis=1)
```

```python
tf.math.reduce_std(small_tensor)
```
