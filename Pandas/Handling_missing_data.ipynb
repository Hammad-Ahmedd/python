## Handling missing data

```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Ali", "Sara", "Hammad", "Usman"],
    "Age": [10, np.nan, 20, np.nan],
    "Score": [50, 70, np.nan, 90]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())
print(df.isnull().sum())
print(df.dropna())
```

```python
df["Age"].fillna(df["Age"].mean())
```

```python
print(df.fillna(method="ffill"))
print(df.fillna(method="bfill"))
```
