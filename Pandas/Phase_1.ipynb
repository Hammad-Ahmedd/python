## Sorting

```python
sortDf = pd.DataFrame({
    "Name": ["Ali", "Sara", "Hammad", "Usman"],
    "Age": [20, 15, 25, 18],
    "Score": [50, 90, 70, 60]
})

print(sortDf)
```

```python
sortDf.sort_values(["Age","Score"],ascending=[True,True])
```
